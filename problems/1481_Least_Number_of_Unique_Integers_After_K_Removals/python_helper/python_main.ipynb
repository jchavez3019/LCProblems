
<div class="elfjS" data-track-load="description_content"><p>Given an array of integers&nbsp;<code>arr</code>&nbsp;and an integer <code>k</code>.&nbsp;Find the <em>least number of unique integers</em>&nbsp;after removing <strong>exactly</strong> <code>k</code> elements<b>.</b></p>

<ol>
</ol>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input: </strong>arr = [5,5,4], k = 1
<strong>Output: </strong>1
<strong>Explanation</strong>: Remove the single 4, only 5 is left.
</pre>
<strong class="example">Example 2:</strong>

<pre><strong>Input: </strong>arr = [4,3,1,1,3,3,2], k = 3
<strong>Output: </strong>2
<strong>Explanation</strong>: Remove 4, 2 and either one of the two 1s or three 3s. 1 and 3 will be left.</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>1 &lt;= arr.length&nbsp;&lt;= 10^5</code></li>
	<li><code>1 &lt;= arr[i] &lt;= 10^9</code></li>
	<li><code>0 &lt;= k&nbsp;&lt;= arr.length</code></li>
</ul></div>


```
from typing import *
from collections import defaultdict
```

```
class Solution:
    def findLeastNumOfUniqueInts(self, arr: List[int], k: int, verbose: bool = False) -> int:
        n = len(arr)
        # get the frequency of each element
        freq = defaultdict(int)
        for a in arr:
            freq[a] += 1

        unique_elements = len(freq) # number of unique elements
        counting_arr = [0] * (n + 1)

        # count the number of unique elements each possible frequency has
        for v in freq.values():
            counting_arr[v] += 1

        for i, c in enumerate(counting_arr):
            if c == 0:
                # no unique elements have a frequency of i
                continue

            # the number of unique elements we can remove is limited by how many times
            # we can remove a unique element with a frequency of i as well as the number
            # of unique elements that do have a frequency of i
            to_remove = min(k // i, c)

            # the number of integers we decrease k by is the number of unique elements
            # we are removing times their frequency
            k_remove = to_remove * i

            if to_remove == 0:
                # if can no longer remove integers at some time step, we cannot again in a
                # future iteration since the frequency is strictly increasing
                break

            # decrement k by the number of integers we have removed
            k -= k_remove
            # decrement the number of unique elements we have removed
            unique_elements -= to_remove

        return unique_elements
def main():
    test_cases = {
        "1": {
            "arr": [5,5,4],
            "k": 1,
            "expected": 1,
        },
        "2": {
            "arr": [4,3,1,1,3,3,2],
            "k": 3,
            "expected": 2,
        },
        "3": {
            "arr": [2,4,1,8,3,5,1,3],
            "k": 3,
            "expected": 3,
        },
    }

    solution = Solution()

    for tk, targs in test_cases.items():
        expected = targs.pop("expected", None)
        ret = solution.findLeastNumOfUniqueInts(**targs, verbose=True)
        if expected is not None:
            passed = ret == expected
        else:
            passed = None
        print(f"test case {tk}: {targs}\nReturned: {ret}, Expected: {expected}\nPassed:{passed}\n")


main()
```
