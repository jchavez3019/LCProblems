
<div class="elfjS" data-track-load="description_content"><p>You are given an integer <code>num</code>. You can swap two digits at most once to get the maximum valued number.</p>

<p>Return <em>the maximum valued number you can get</em>.</p>

<p>&nbsp;</p>
<p><strong class="example">Example 1:</strong></p>

<pre><strong>Input:</strong> num = 2736
<strong>Output:</strong> 7236
<strong>Explanation:</strong> Swap the number 2 and the number 7.
</pre>

<p><strong class="example">Example 2:</strong></p>

<pre><strong>Input:</strong> num = 9973
<strong>Output:</strong> 9973
<strong>Explanation:</strong> No swap.
</pre>

<p>&nbsp;</p>
<p><strong>Constraints:</strong></p>

<ul>
	<li><code>0 &lt;= num &lt;= 10<sup>8</sup></code></li>
</ul>
</div>


```
from typing import *
```

```

class Solution:
    def maximumSwap(self, num: int) -> int:
        # return self.naive_solution(num)
        return self.sorting_args(num)

    def naive_solution(self, num: int) -> int:
        """
        This is a naive solution that runs in O(n^2)
        """
        num_str = str(num)
        n = len(num_str)
        largest_num = num
        for i in range(n-1):
            for j in range(i+1, n):
                new_num_str = num_str[:i] + num_str[j] + num_str[i+1:j] + num_str[i] + num_str[j+1:]
                largest_num = max(largest_num, int(new_num_str))

        return largest_num

    def sorting_args(self, num: int) -> int:
        """
        This solution runs in O(n).

        We first do a backward pass where we track the largest element in the range [i,n].
        Afterwards, we do a forward pass where we try swapping with the largest elements
        from the forward pass in order to find the largest number.
        """
        num_str = str(num)
        num_list = [int(s) for s in num_str]
        n = len(num_list)

        # largest_subseq will hold the index in the num_list that
        # has the largest value in the range [i,n]
        largest_subseq = [0] * n
        max_num, max_idx = -1, -1
        for i in range(n-1, -1, -1):
            curr = num_list[i]
            if curr > max_num:
                max_num = curr
                max_idx = i
            largest_subseq[i] = max_idx

        largest_num = num
        for i in range(n):
            j = largest_subseq[i]
            if i == j:
                # no point swapping an element with itself
                continue
            # perform a swap in the string with indices i and j
            new_num_str = num_str[:i] + num_str[j] + num_str[i+1:j] + num_str[i] + num_str[j+1:]
            largest_num = max(largest_num, int(new_num_str))

        return largest_num

def main():
    test_cases = {
        "1": {
            "nums": 2736,
            "expected": 7236,
        },
        "2": {
            "num": 9973,
            "expected": 9973,
        },
        "3": {
            "num": 1993,
            "expected": 9913,
        }
    }

    solution = Solution()

    for tk, targs in test_cases.items():
        expected = targs.pop("expected", None)
        ret = solution.isZeroArray(**targs, verbose=True)
        if expected is not None:
            passed = ret == expected
        else:
            passed = None
        print(f"test case {tk}: {targs}\nReturned: {ret}, Expected: {expected}\nPassed:{passed}\n")

main()
```
